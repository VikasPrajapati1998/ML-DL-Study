# **Matplotlib**
- **Data Visualization** : Data visualization is a quick, easy way to convey concepts in a universal manner.
- **Plotting Libraries** : /
    **Matplotlib :** low level, provides lots of freedom \
    **Pandas Visualization :** easy to use interface, build on Matplotlib \
    **Seaborn :** high-level interface, great default styles \
    **ggplot :** based on R's ggplot2, uses Grammar of Graphics \
    **Plotly :** can create interactive plots

## **Matplotlib Graphs**
- **Basic**
    - plot(x, y)
    - scatter(x, y)
    - bar(x, height)
    - stem(x, y)
    - step(x, y)
    - fill_between(x, y1, y2)
    - stackplot(x, Y)
- **Array Plot**
    - imshow(arr1)
    - pcolormesh(a, b, c)
    - contour(a, b, c)
    - contourf(a, b, c)
    - barbs(x, y, u, v)
    - quiver(x, y, u, v)
    - streamplot(x, y, u, v)
- **Statistics Plot**
    - hist(x)
    - boxplot(x)
    - violinplot(x)
    - eventplot(x)
    - hist2d(x, y)
    - hexbin(x, y,C)
    - pie(x)
- **Unstructured Coordinates**
    - tricontour(x, y, z)
    - tricoutourf(x, y, z)
    - tripcolor(x, y, z)
    - triplot(x, y)
- **3D**
    - 3D scatterplot
    - 3D surface
    - Triangular 3D surfaces
    - 3D voxel / volumetric plot
    - 3D wireframe plot

```python
x = [x**2 for x in range(-20, 21)]
y = [y for y in range(-20, 21)]
plt.plot(x, y, color="r", label="Parabola")

plt.title("Parabola", fontsize=20, color="blue")
plt.xlabel("x-axis", fontsize=12, color="brown")
plt.ylabel("y-axis", fontsize=12, color="brown")
           
plt.legend()
plt.show()
```

- bins=list / 'auto'
- histtype=step / stepfilled / bar / barstack

```python
bin_lst = [x*10 for x in range(1, 9)]
plt.hist(x, color="green", edgecolor="blue", linewidth=3, bins="auto")
plt.title("Random", fontsize=20)
plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.show()
```

```python
bin_lst = [x*10 for x in range(1, 9)]
plt.hist(x, "auto", (10, 200), color="pink", edgecolor="k", linewidth=3) # hist(x, bins, range)
plt.title("Random", fontsize=20)
plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.show()
```

```python
bin_lst = [x*10 for x in range(1, 9)]
plt.hist(x, "auto", color="green", edgecolor="k", width=8, linewidth=3, bottom=10, align="mid", histtype="bar") # hist(x, bins, range)
plt.title("Random", fontsize=20)
plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.show()
```

```python
bin_lst = [x*10 for x in range(1, 9)]
plt.hist(x, "auto", color="green", edgecolor="k", linewidth=3, bottom=10, align="mid", histtype="step") # hist(x, bins, range)
plt.title("Random", fontsize=20)
plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.show()
```

```python
bin_lst = [x*10 for x in range(1, 9)]
plt.hist(x, "auto", color="green", height=8, edgecolor="k", linewidth=3, bottom=10, align="mid", histtype="bar", orientation="horizontal") # orientation=vertical/horizontal
plt.title("Random", fontsize=20)
plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.show()
```

```python
# make a line in mid of histogram
x = np.random.randint(10, 90, 60)
bin_lst = [x*10 for x in range(1, 9)]
plt.hist(x, color="red", edgecolor="g", linewidth=2, label="python")
plt.title("Random", fontsize=20)
plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.axvline(45, color="k", linewidth=5, label="Mid Data") # vertical line
plt.axhline(4, color="cyan", linewidth=5, label="Half Data") # horizontal line

plt.legend()
plt.grid()
plt.show()
```

```python
# widths=n, labels=[]
# patch_artist=True/False
x = [x*10 for x in range(1, 11)]
plt.boxplot(x, widths=0.5, labels=["Python"], patch_artist=True)
plt.show()
```

```python
# showmeans=True/False
x = [x*10 for x in range(1, 11)]
plt.boxplot(x, labels=["python"], showmeans=True)
plt.show()
```

```python
# showmeans=True/False
x = [x*10 for x in range(1, 9)]
x.append(150)
print(x)
plt.boxplot(x, labels=["python"], showmeans=True)
plt.show()
```

```python
# whis=2
x = [x*10 for x in range(1, 9)]
x.append(150)
print(x)
plt.boxplot(x, labels=["python"], showmeans=True, whis=2)
plt.show()
```

```python
# whis=2
# sym="color and shape
x = [x*10 for x in range(1, 9)]
x.append(150)
print(x)
plt.boxplot(x, labels=["python"], showmeans=True, sym="r*")
plt.show()
```

```python
# boxprops=dict(), and capprops=dict()
x = [x*10 for x in range(1, 9)]
print(x)
plt.boxplot(x, labels=["python"], showmeans=True, sym='r^', boxprops=dict(color='b'), capprops=dict(color='cyan'), whiskerprops=dict(color="m"))
plt.show()
```

- **ylabel** : Add a label to the Axes's y-axis.
