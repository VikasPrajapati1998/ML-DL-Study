# **Series**
- It is defined as a one-dimensional array that is capable of storing various data types.
- s = pd.Series(data, index, dtype, name)
- data(list or dict or array), index(list), dtype(int, float, complex, bool), name(str)
- Length of data and index could be different.

```python
lst = [x for x in range(3, 9)]
s = pd.Series(lst)
print(s)
print(type(s))

# access data using index
print(s[2])
```

```python
lst = [x for x in range(3, 9)]
idx = [chr(x) for x in range(ord('a'), ord('g'))]
s = pd.Series(lst, index=idx, dtype="float", name="numbers")
print(s)
print(type(s))
```

```python
dct = {x : x*2 for x in range(1, 9)}
s = pd.Series(dct)
print(s)
```

```python
dct = {chr(96+x) : x*2 for x in range(1, 10)}
print(dct); print()
s = pd.Series(dct)
print(s)
```

```python
# different length of data and index
dct = {chr(k) : k-96 for k in range(ord('a'), ord('g'))}
print(dct)
s = pd.Series(dct)
print(s)
print()

idx = [chr(x) for x in range(ord('a'), ord('d'))]
print(idx)
s = pd.Series(dct, index=idx)
print(s)
print()

idx = [chr(x) for x in range(ord('a'), ord('m'))]
print(idx)
s = pd.Series(dct, index=idx)
print(s)
print()

idx = [chr(x) for x in range(ord('A'), ord('M'))]
print(idx)
s = pd.Series(dct, index=idx)
print(s)
print()
```

```python
dic = {"name" : ['python', 'c', 'c++', 'java'], "por" : [12, 13, 14, 15], "rank" : [1, 4, 3, 2]}
s = pd.Series(dic)
print(s)
print(s.dtype)
print(type(s))
```

## **Arithmetic Operations in Python Pandas**

```python
df = pd.DataFrame({'A': [12, 21, 34, 47], 'B': [5, 6, 7, 8]})
print(df); print()

# Addition
df['Add'] = df['A'] + df['B']
print(df); print()

# Substraction
df['Sub'] = df['A'] - df['B']
print(df); print()

# Multiply
df['Mult'] = df['A'] * df['B']
print(df); print()

# Division
df['Div'] = df['A'] // df['B']
print(df); print()

# Division
df['Divi'] = df['A'] / df['B']
print(df); print()

# Remainder
df['Rem'] = df['A'] % df['B']
print(df); print()

# Exponent
df['Expo'] = df['A'] ** df['B']
print(df); print()

df
```

```python
df = pd.DataFrame({'A': [x*11 for x in range(1, 10)], 'B': [x for x in range(1, 10)], 'C': [x*13 for x in range(1, 10)]})
print(df); print()

df["Python"] = df['A']%2 == 0
df["Java"] = df['B'] >= 5
df["Cpp"] = ~df['Python']
print(df)
print()

df["Shell"] = np.random.randint(1, 10, 9)
print(df)
print()

df
```

## **Delete and Insert Functions in Python Pandas**

### **Insert**
- **df.insert(*column_index, name_of_column, data*)**

```python
df = pd.DataFrame({'A': [x*11 for x in range(1, 10)], 'B': [x for x in range(1, 10)], 'C': [x*13 for x in range(1, 10)]})
print(df); print()

# insert a column
data = [x for x in range(1, 10)]
df.insert(0, "idx", data)
print(df); print()

data = [x**2 for x in range(1, 10)]
df.insert(2, "sqrt", data)
print(df); print()

df.insert(4, "copy_of_A", df['A'])
print(df)
df
```

```python
df = pd.DataFrame({'A': [x*11 for x in range(1, 10)], 'B': [x for x in range(1, 10)], 'C': [x*13 for x in range(1, 10)]})
print(df); print()
df["Python"] = df['A'][2:7]
print(df)
df
```

### **Delete**
- **df.pop(*column_name*)**
- **del**

```python
df = pd.DataFrame({'A': [x*11 for x in range(1, 10)], 'B': [x for x in range(1, 10)], 'C': [x*13 for x in range(1, 10)]})
print(df); print()
p = df.pop('B') # pop return deleted column
print(p); print()
print(df); print()
df
```

```python
df = pd.DataFrame({'A': [x*11 for x in range(1, 10)], 'B': [x for x in range(1, 10)], 'C': [x*13 for x in range(1, 10)]})
print(df); print()
del df['A'] # del does not return deleted column
print(df)
print()
df
```

### ***dropna()***
- **df.dropna(axis=1)** : column drop
- **df.dropna(axis=0)** : row drop
- **df.dropna(how=all)** : drop that row, which have all value NaN \
    **how=all, axis=0** : drop that row, which have all value NaN \
    **how=all, axis=1** : drop that column, which have all value NaN
- **df.dropna(how=any)** : drop that row, which have any value NaN \
    **how=any, axis=0** : drop that row, which have any value NaN \
    **how=any, axis=1** : drop that column, which have any value NaN
- **df.dropna(subset=[columnNames])** : drop that row in which columnNames have NaN
- **df.dropna(thresh=n)** : drop that row, which have `n` NaN value
- **df.dropna(inplace=True/False)** : change in actual data frame

```python
# dropna()
ndf = df.dropna()
ndf.isnull().sum()
```

```python
ndf = df.dropna(axis=1) # drop column = axis=1
ndf.isnull().sum()
```

```python
ndf = df.dropna(axis=0) # drop row - axis=0
ndf.isnull().sum()
```

```python
ndf = ndf.isnull().describe()
ndf
```

```python
# how=any : drop that row if have any NaN value
ndf = df.dropna(how="any")
ndf.isnull().sum()
```

```python
# how=any, axis=0 : drop that row if have any NaN value
ndf = df.dropna(how="any", axis=0)
ndf.isnull().sum()
```

```python
# how=any, axis=1 : drop that column if have any NaN value
ndf = df.dropna(how="any", axis=1)
ndf.isnull().sum()
```

```python
# how=all : drop the rows that's all value NaN
ndf = df.dropna(how="all")
ndf.isnull().sum()
```

```python
# how=all, axis=0 : drop the rows that's all value NaN
ndf = df.dropna(how="all", axis=0)
ndf.isnull().sum()
```

```python
# how=all, axis=1 : drop the coloumn that's all value NaN
ndf = df.dropna(how="all", axis=1)
ndf.isnull().sum()
```

### ***fillna()***
- **df.fillna(n)** : `NaN` fill with `n`
- **df.fillna({columnName1 : value1, columnName2 : value2})** : `NaN` fill with of _columnName1_ with _value1_ and _columnName2_ with _value2_
- **df.fillna(method=ffill/bfill)** : ffill : forward fill. Take upper value `&` bfill : backward fill. Take lower value.
