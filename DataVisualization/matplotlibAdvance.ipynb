```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right-top'],['left', 'right_bottom']])
```

```python
data = np.array([12, 15, 22, 34, 16, 45, 26, 22, 39, 37, 4, 43, 29, 18, 21, 45, 28, 8, 41, 7, 5, 35, 47, 19])
data
```

```python
fig = plt.figure(1)  # an empty figure with no Axes
plt.plot(data)

fig, ax = plt.subplots(1)  # a figure with a single Axes
ax.plot(data, color='r')

plt.show()
```

```python
x = np.arange(0, 99)
y = x**2

fig, axs = plt.subplots(1, 2)  # a figure with a 1x1 grid of Axes
axs[0].plot(x, y, color='g', linewidth=3.5)
axs[1].plot(x, y, color='m', linewidth=3.5)

plt.show()
```

```python
x = np.arange(0, 99)
y = x**2

fig, axs = plt.subplots(2, 2, figsize=(10, 5))
fig.suptitle('Squared Values of X', fontsize=16)

axs[0, 0].plot(x, y, color='g', linewidth=3.5)
axs[0, 0].set_title('Single Plot', fontsize=14)
axs[0, 0].set_xlabel('X', fontsize=12)
axs[0, 0].set_ylabel('X Squared', fontsize=12)

axs[0, 1].plot(x, y, color='m', linewidth=3.5)
axs[0, 1].set_title('Two Plots', fontsize=14)
axs[0, 1].set_xlabel('X', fontsize=12)
axs[0, 1].set_ylabel('X Squared', fontsize=12)

axs[1, 0].scatter(x, y, color='g', s=3.5)
axs[1, 0].set_title('Scatter Plot', fontsize=14)
axs[1, 0].set_xlabel('X', fontsize=12)
axs[1, 0].set_ylabel('X Squared', fontsize=12)

axs[1, 1].scatter(x, y, color='m', s=3.5)
axs[1, 1].set_title('Another Scatter Plot', fontsize=14)
axs[1, 1].set_xlabel('X', fontsize=12)
axs[1, 1].set_ylabel('X Squared', fontsize=12)

plt.show()
```

```python
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right-top'],['left', 'right_bottom']])
```

```python
x = np.arange(0, 99)
y = x**2

fig, axs = plt.subplot_mosaic([['left', 'right-top'],['left', 'right-bottom']])
fig.suptitle('Squared Values of X', fontsize=16)

axs['left'].plot(x, y, color='g', linewidth=3.5)
axs['left'].set_title('Single Plot', fontsize=14)
axs['left'].set_xlabel('X', fontsize=12)
axs['left'].set_ylabel('X Squared', fontsize=12)

axs['right-top'].plot(x, y, color='m', linewidth=3.5)
axs['right-top'].set_title('Two Plots', fontsize=14)
axs['right-top'].set_xlabel('X', fontsize=12)
axs['right-top'].set_ylabel('X Squared', fontsize=12)

axs['right-bottom'].plot(x, y, color='r', linewidth=3.5)
axs['right-bottom'].set_title('Two Plots', fontsize=14)
axs['right-bottom'].set_xlabel('X', fontsize=12)
axs['right-bottom'].set_ylabel('X Squared', fontsize=12)

plt.show()
```
